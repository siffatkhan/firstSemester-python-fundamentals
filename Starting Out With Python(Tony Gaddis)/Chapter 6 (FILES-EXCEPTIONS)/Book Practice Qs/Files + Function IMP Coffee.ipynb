```python
def append():      
    with open("coffee.txt","a") as f:
        another="y"
        while another=="y" or another=="Y":
            coffee_name=input("Enter Coffee Brand: ")
            quantity=input("Enter Coffee Quantity: ")
            f.write(f"{coffee_name}\n")
            f.write(f"{quantity}\n")
            another=input("\nDo you want to add another record: Y OR N: ")
            print()
        print("\nRecords are added to the coffee.txt file.")
append()
```

```python
def read():      
    with open("coffee.txt","r") as f:
        first=f.readline()
        while first!="":
            first=first.rstrip("\n")
            print(f"Coffee Brand: {first}")
            quantity=f.readline()
            quantity=quantity.rstrip("\n")
            first=f.readline()
            print(f"Coffee Quantity: {quantity}")
read()
```

```python
import os      
def update():
    found=False
    search = input('Enter a description to search for: ')
    new_qty = int(input('Enter the new quantity: '))
    with open('coffee.txt', 'r') as coffee_file:
        with open('temp.txt', 'w') as temp_file:
            first=coffee_file.readline()
            while first!="":
                qty = float(coffee_file.readline())
                first=first.rstrip("\n")
                if first==search:
                    temp_file.write(first + '\n') 
                    temp_file.write(str(new_qty) + '\n')
                    found=True
                else:
                    temp_file.write(first + '\n')
                    temp_file.write(str(qty) + '\n')
            
                first=coffee_file.readline()
            os.remove('coffee.txt')
            os.rename('temp.txt', 'coffee.txt')
            if found:
                print('The file has been updated.') 
            else:
                print('That record was not found in the file.')

update()
```

```python
import os 
def delete():
    found=False
    search = input('Which coffee do you want to delete? ')
    with open("coffee.txt","r") as coffee_file:
        with open('temp.txt', 'w') as temp_file:
            first=coffee_file.readline()
            while first!="":
                qty = float(coffee_file.readline())
                first = first.rstrip('\n')
                if first!=search:
                    temp_file.write(first + '\n')
                    temp_file.write(str(qty) + '\n')
                else:
                    found=True
                first=coffee_file.readline()
    os.remove("coffee.txt")
    os.rename('temp.txt', 'coffee.txt')
    if found:
        print('The file has been updated.')
    else:
        print('That item was not found in the file.')
delete()
```
