```python
import tkinter as tk

class CalculatorApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Custom Shaped Calculator")


        self.master.configure(bg='black')


        self.canvas = tk.Canvas(master, bg='black', highlightthickness=0)
        self.canvas.pack(fill=tk.BOTH, expand=True)


        self.entry = tk.Entry(self.canvas, font=('Helvetica', 24), bd=10, insertwidth=4, justify='right', bg='black', fg='white')
        self.canvas.create_window(300, 40, window=self.entry, anchor='n')


        buttons = [
            ('AC', 100, 100, 'maroon'), ('÷', 200, 100, 'black'), ('*', 300, 100, 'black'), ('ON', 400, 100, 'maroon'),
            ('7', 100, 220, 'black'), ('8', 200, 220, 'black'), ('9', 300, 220, 'black'), ('-', 400, 220, 'maroon'),
            ('4', 100, 340, 'black'), ('5', 200, 340, 'black'), ('6', 300, 340, 'black'), ('+', 400, 340, 'maroon'),
            ('1', 100, 460, 'black'), ('2', 200, 460, 'black'), ('3', 300, 460, 'black'), ('=', 400, 460, 'maroon'),
            ('0', 100, 580, 'black'), ('.', 300, 580, 'black')
        ]


        for (text, x, y, color) in buttons:
            if text == '0':
                button = tk.Button(self.canvas, text=text, font=('Helvetica', 18), padx=25, pady=20, command=lambda t=text: self.button_click(t), bg=color, fg='white')
                self.canvas.create_rectangle(x, y, x + 200, y + 100, fill=color, outline='black', width=2)
                self.canvas.create_window(x + 100, y + 50, window=button, anchor='center')
            elif text == 'AC':
                button = tk.Button(self.canvas, text=text, font=('Helvetica', 18), padx=20, pady=20, command=lambda: self.reset_input(), bg=color, fg='white')
                self.canvas.create_rectangle(x, y, x + 100, y + 100, fill=color, outline='black', width=2)
                self.canvas.create_window(x + 50, y + 50, window=button, anchor='center')
            elif text == 'ON':
                button = tk.Button(self.canvas, text=text, font=('Helvetica', 18), padx=20, pady=20, command=lambda: self.clear_input(), bg=color, fg='white')
                self.canvas.create_rectangle(x, y, x + 100, y + 100, fill=color, outline='black', width=2)
                self.canvas.create_window(x + 50, y + 50, window=button, anchor='center')
            else:
                button = tk.Button(self.canvas, text=text, font=('Helvetica', 18), padx=20, pady=20, command=lambda t=text: self.button_click(t), bg=color, fg='white')
                self.canvas.create_rectangle(x, y, x + 100, y + 100, fill=color, outline='black', width=2)
                self.canvas.create_window(x + 50, y + 50, window=button, anchor='center')

    def button_click(self, value):
        current = self.entry.get()
        if value == '=':
            try:
                result = eval(current)
                self.entry.delete(0, tk.END)
                self.entry.insert(0, str(result))
            except Exception as e:
                self.entry.delete(0, tk.END)
                self.entry.insert(0, "Error")
        else:
            self.entry.delete(0, tk.END)
            self.entry.insert(0, current + value)

    def reset_input(self):
        current = self.entry.get()
        if current:
            self.entry.delete(len(current) - 1, tk.END)

    def clear_input(self):
        self.entry.delete(0, tk.END)

if __name__ == "__main__":
    root = tk.Tk()
    app = CalculatorApp(root)
    root.mainloop()
```
