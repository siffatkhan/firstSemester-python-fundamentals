```python
print("""\t   0. Create New List
           1. Add items to the list
           2. Insert items at any specified index
           3. Append item to list
           4. Display the list items
           5. Remove the specific items from the list
           6. Sort the items of the list
           7. Find item in the list
           8. Reverse items of the list
           9. Min value of the list
          10. Max valie of the list
          11. Delete item from the list
          12. Quit the program""")
print("======================================================================================")
while True:
    option=int(input("\nEnter The No Of Function U Want To Perform."))
    list1=["IMS","BCS","B","2023-2027"]
    if option>0 and option<12 and option!=4:
        print(list1)
    if option==0:
        print("New List Created: ",list1)
        print("======================================================================================")
    elif option==1:
        x=int(input("\nHow Much Values U Wanna Add To List"))
        for _ in range(1,x+1):
            value=input(f"Enter The {_} Value To Add In List: ")
            list1.append(value)
        print(f"\nYour New List Is After Adding Value: {list1}")
        print("======================================================================================")
    elif option==2:
        value=input(f"\nEnter The Value To Insert In List: ")
        index=int(input("On what Index U Want To Insert: "))
        list1.insert(index,value)
        print(f"\nYour New List After Inserting Values: {list1}")
        print("======================================================================================")
    elif option==3:
        value=input(f"\nEnter The Value To Append In List: ")
        list1.append(value)
        print(f"\nYour New List After Appending: {list1}")
        print("======================================================================================")
    elif option==4:
        print("\n",list1)
        print("======================================================================================")
    elif option==5:
        value=input(f"\nEnter Value To Remove From List: ")
        list1.remove(value)
        print(f"\nYour New List After Removal: {list1}")
        print("======================================================================================")
    elif option==6:
        list1.sort()
        print(f"\nYour New List After Sorting: {list1}")
        print("======================================================================================")
    elif option==7:
        try:
            while True:
                value=input(f"\nEnter Findng Item In List: ")
                z=list1.index(value)
                print("Value is at",z,"index.")
                break
        except:
            print("Value is not in list.")
        print("======================================================================================")
    elif option==8:
        list1.reverse()
        print(f"\nYour New List After Reversing: {list1}")
        print("======================================================================================")
    elif option==9:
        print(f"\nThe Min Value Of List Is: {min(list1)}")
        print("======================================================================================")
    elif option==10:
        print(f"\nThe Max Value Of List Is: {max(list1)}")
        print("======================================================================================")
    elif option==11:
        value=int(input(f"\nWhat Index Value U Want To Remove."))
        del list1[value]
        print(f"\nYour New List After Deletion: {list1}")
        print("======================================================================================")
    elif option==12:
        print("\nProgram Quitted")
        print("======================================================================================")
        break
    else:
        raise ValueError ("Invalid Input: Enter Correct Option")
```

```python
print("""\t   0. Create New List
           1. Add items to the list
           2. Insert items at any specified index
           3. Append item to list
           4. Display the list items
           5. Remove the specific items from the list
           6. Sort the items of the list
           7. Find item in the list
           8. Reverse items of the list
           9. Min value of the list
          10. Max valie of the list
          11. Delete item from the list
          12. Quit the program""")

while True:
    choose=int(input("Enter fUNCTION NO."))
    list_z=["KALEEM","BASIR","ALI","SAAD"]
    if choose>0 and choose<12 and choose!=4:
        print(list_z)
    if choose==0:
        print("List Created: ",list_z)
    elif choose==1:
        x=int(input("Values U Wanna Add To List"))
        for z in range(1,x+1):
            value=input(f"Enter The {z} Value To Add In List: ")
            list_z.append(value)
        print(f"Adding Value: {list_z}")
    elif choose==2:
        value=input(f"Enter Insert value: ")
        index=int(input("Insert INDEX: "))
        list_z.insert(index,value)
        print(f"Inserting Values: {list_z}")
    elif choose==3:
        value=input(f"Enter Appending value: ")
        list_z.append(value)
        print(f"Appending: {list_z}")

    elif choose==4:
        print(list_z)
     
    elif choose==5:
        value=input(f"Enter Value To Remove: ")
        list_z.remove(value)
        print(f"Removal: {list_z}")
       
    elif choose==6:
        list_z.sort()
        print(f"Sorting: {list_z}")
    
    elif choose==7:
        try:
            while True:
                value=input(f"Enter Findngt: ")
                z=list_z.index(value)
                print("Value is at",z,"index.")
                break
        except:
            print("Value is not in list.")
        
    elif choose==8:
        list_z.reverse()
        print(f"Reversing: {list_z}")
       
    elif choose==9:
        print(f" Min Value: {min(list_z)}")
      
    elif choose==10:
        print(f"Max Value: {max(list_z)}")
        
    elif choose==11:
        value=int(input("Index To Remove."))
        del list_z[value]
        print(f"Delete: {list_z}")
       
    elif choose==12:
        print("Quit")
        break
```
