```python
# MAIN FUNCTION TO BE RUN :--

room=int(input("Enter No Of Rooms"))        
while room<1:                             #VALIDATION THROUGH LOOP
    print("INVALID ROOM INPUT")
    print("-------------------------")
    room=int(input("ENTER CORRECT NO OF ROOMS"))
tileSize=int(input("Enter Size Of Tiles In Inches"))
while tileSize<1 or tileSize>12:
    print("INVALID TILE INPUT")
    print("-------------------------")
    tileSize=int(input("ENTER CORRECT SIZE OF TILES IN INCHES "))
def main():
    calc() 
main()
```

```python
# ADDITIONAL FUNCTIONS INSIDE MAIN FUNCTION :--


# FUNCTION FOR THE LENGTH CALCULATION :--
#------------------------------------------
def additionlength(product):
    if product%tileSize!=0:
        addition=(product//tileSize)+1
        return addition
    else:
        addition=(product//tileSize)
        return addition

    
## FUNCTION FOR THE WIDTH CALCULATION :--
#------------------------------------------
def additionwidth(product_wid):
    if product_wid%tileSize!=0:
        addwid=(product_wid//tileSize)+1
        return addwid
    else:
        addwid=(product_wid//tileSize)
        return addwid

    
# FUNCTION FOR THE TOTAL TILES CALCULATION :--
#------------------------------------------
def tileboxes(total):
    if total%20==0:
        print("---------------------------------------------")
        print(f"\nNumber of Boxes needed: {total/20}")
        print(f"Number of Tiles left: {total%20}")
        print("---------------------------------------------")
    elif total%20!=0:
        i=(total//20)+1
        print("---------------------------------------------")
        print(f"\nNumber of Boxes needed: {i}")
        print(f"Number of Tiles left: {(i*20)-total}")
        print("---------------------------------------------")
        
```

```python
# ALL THE CALCULATION & MULTIPLICATION OF PROBLEM 1 :--

def calc():
        total=0
        count=1
        while count<=room:
            length=input(f"\nEnter {count} Room Length: ")    # LENGTH INPUT
            if " " in length:
                new_length=length.split()
                new_length1=len(new_length)
                if new_length1==1:
                    feetL=int(new_length[0])
                    product=feetL*12
                    x=additionlength(product)        #VALIDATION
                else:
                    feetL=int(new_length[0])
                    inchL=int(new_length[1])
                    if inchL>12 or inchL<0:
                        raise Exception("INVALID INCHES INPUT")
                    product=(feetL*12)+inchL
                    x=additionlength(product)      
            elif " " not in length:
                length=int(length)
                product=length*12
                x=additionlength(product)
                
            width=input("Enter Room Width: ")         #WIDTH INPUT 
            if " " in width:
                new_width=width.split()
                new_width1=len(new_width)
                if new_width1==1:
                    feetW=int(new_width[0])
                    product_wid=feetW*12
                    y=additionwidth(product_wid)         #VALIDATION
                else:
                    feetW=int(new_width[0])
                    inchW=int(new_width[1])
                    if inchW>12 or inchW<0:
                        raise Exception("INVALID INCHES INPUT")
                    product_wid=(feetW*12)+inchW
                    y=additionwidth(product_wid)
            elif " " not in width:
                width=int(width)
                product_wid=width*12
                y=additionwidth(product_wid) 
            z=x*y
            print(f"\n{count} Room Require {z} Tiles\n")
            total+=z
            count+=1
        print("---------------------------------------------")
        print(f"\nTotal Tiles Required Are {total}\n")
        tileboxes(total)
```

```python
# MAIN FUNCTION FOR PROBLEM 2 TO BE RUN :--

def main():
    while True:
        choice=int(input("""->> What Are U Looking For:\n
-> PRESS 1 For LOOKING PHONE NUMBER
-> PRESS 2 For LOOKING ADDRESS
-> PRESS 3 For LOOKING BOTH
-> PRESS 4 For QUITING 
========================================================: """))
        if choice==1:         # CHOICES GIVEN FOR REQUIRED SEARCHES
            phoneno()
            
        elif choice==2:
            address() 
            
        elif choice==3:
            both()
            
        elif choice==4:
            print("THANK YOU For Using Our Program")
            print("========================================================")
            break
        else:
            print("--------------------------------------------------------")
            print("INVALID INPUT TRY AGAIN")
            print("--------------------------------------------------------")
            choice=int(input("""\nWhat Are U Looking For:\n
-> PRESS 1 For LOOKING PHONE NUMBER
-> PRESS 2 For LOOKING ADDRESS
-> PRESS 3 For LOOKING BOTH
-> PRESS 4 For QUITING 
========================================================: """))
main()
```

```python
 # FUNCTION FOR LOOKING PHONE NUMBER :--

def phoneno():
    with open("phonedir1.txt","r") as f:
        search=input("\nENTER FIRST NAME: ")+" "+input("ENTER LAST NAME: ")
        search=search.upper()
        flag=False
        for x in f:                 # LOGIC OF LIST SPLITING
            y=x.rstrip("\n")
            list1=y.split()
            name1=list1[0]+" "+list1[1]  
            if search == name1:
                print("--------------------------------------------------------")
                print(f"\n^-^ The PHONE NUMBER Of {search} Is:-> {list1[-1]}")
                print("\n--------------------------------------------------------")
                flag=True
                break
        if flag==False:
            print("--------------------------------------------------------")
            print("\n--> NO SUCH PERSON NAME IS PRESENT IN FILE.")
            print("\n--------------------------------------------------------")
        repeat()         
```

```python
# FUNCTION FOR LOOKING ADDRESS :--

def address():
    with open("phoneextension.txt","r") as f:
        search=input("\nENTER FIRST NAME: ")+" "+input("ENTER LAST NAME: ")
        search=search.upper()
        print()
        flag=False
        end=f.readline()
        while end!="":                  #LOGIC OF LIST SPLITTING
            end=end.rstrip("\n")
            list1=end.split()
            name1=list1[0]+" "+list1[1]
            end=f.readline()
            if search == name1:
                print("--------------------------------------------------------")
                print(f"\n^-^ The ADDRESS Of {search} is:->  {end}")
                print("--------------------------------------------------------")
                flag=True
                break
            end=f.readline()
        if flag==False:
            print("--------------------------------------------------------")
            print("\n--> NO SUCH PERSON NAME IS PRESENT IN FILE")
            print("\n--------------------------------------------------------")
# address()
```

```python
# FUNCTION FOR LOOKING ADDRESS AND PHONE NUMBER IN ONE GO :--

def both():
    with open("phoneextension.txt","r") as f:
        search=input("\nENTER FIRST NAME: ")+" "+input("ENTER LAST NAME: ")
        search=search.upper()
        flag=False
        end=f.readline()
        while end!="":                   #LOGIC OF LIST SPLITTING AND INDEXING
            list1=end.split()
            name1=list1[0]+" "+list1[1]
            end=end.rstrip("\n")   
            end=f.readline()
            if search==name1:
                print("--------------------------------------------------------")
                print(f"\n^-^ The PHONE NUMBER Of {search} Is:->  {list1[-1]}") 
                print(f"\n^-^ The ADDRESS Of {search} Is:->  {end}")
                print("--------------------------------------------------------")
                flag=True
                break
            end=f.readline()
        if flag==False:
            print("--------------------------------------------------------")
            print("\n--> NO SUCH PERSON NAME IS PRESENT IN FILE")
            print("\n--------------------------------------------------------")
# both()
```

```python
# FUNCTION MADE FOR THE REPETITION OF PROGRAM :---

def repeat():
    another=input("\nDO YOU WANT TO REPEAT-> YES OR NO: ")
#     another=input()
    if another=="yes" or another=="YES":
        phoneno()
    elif another=="no" or another=="NO":
        print("--------------------------------------------------------")
        print("\n^-^ THANK FOR USING THIS.\n")
        print("--------------------------------------------------------")
    else:
        while True:
            print("\nEnter Correct Command.")
            another=input("YES OR NO")
            if another=="yes" or another=="YES":
                phoneno()
                break
            elif another=="no" or another=="NO":
                print("--------------------------------------------------------")
                print("\n^-^ THANK FOR USING THIS.\n")
                print("--------------------------------------------------------")
                break
            else:
                continue
```

```python
# MAIN FUNCTION FOR THE PROBLEM 3 :--

def main():
    with open("exam.txt","r")as f:
        with open("score.txt","w")as g:
            key=f.readline()
            key=key.rstrip("\n")
            answer=f.readline()  
            while answer!="":
                answer=answer.rstrip("\n")
                anslist=answer.split()
                id1=anslist[0]
                mcqs=anslist[-1]
                correct_total=scoring(key,mcqs,id1)
                writing(correct_total,id1)
                answer=f.readline()
main()
```

```python
#FUNCTION FOR READING DATA FROM THE FILE :--

def scoring(key,exam,id1):
    x=len(key)
    y=len(exam)

    if x>y:                            #LOGIC FOR SCORE DETCTION
        total="Too few answers."
        return print(total)
    elif x<y:
        total="Too many answers."
        return print(total)
    elif x==y:
        total=0
        for index,i in enumerate(exam):
            a=key[index]
            if ord(i)>ord("f"):
                total="Invalid Answer."
                return print(total)
                break 
            elif i == a and ord(i)<ord("g"):
                total+=1 
            elif i != a and ord(a)<ord("g"):
                total+=0
        return print(total)
```

```python
# FUNCTION FOR WRITING SCORES TO THE FILE :--

def writing(score,id1):
    with open("score.txt","w")as q:
        q.write("ID\t"+"Score"+"\n")
        for i in range(6):
            q.write(id1+" ")
            q.write(str(score)+"\n")
```
