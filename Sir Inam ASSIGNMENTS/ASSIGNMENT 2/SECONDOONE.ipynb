```python
def writing():
    with open("phonedir.txt","w") as g:
        number=int(input("How Many People Data U Want To Enter: "))
        for x in range(1,number+1):
            first=input(f"\nEnter {x} Person FIRST NAME   : ")
            last=input(f"Enter {x} Person LAST NAME    : ")
            phone=input(f"Enter {x} Person PHONE NUMBER : ")
            print()
            g.write(first.capitalize()+" ")
            g.write(last.lower()+" ")
            g.write(phone+"\n")
# writing()
            
```

```python
def writingboth():
    with open("phoneext.txt","w") as g:
        number=int(input("How Many People Data U Want To Enter: "))
        for x in range(1,number+1):
            first=input(f"\nEnter {x} Person FIRST NAME   : ")
            last=input(f"Enter {x} Person LAST NAME    : ")
            phone=input(f"Enter {x} Person PHONE NUMBER : ")
            address=input(f"Enter {x} Person HOME ADDRESS: ")
            print()
            g.write(first.capitalize()+" ")
            g.write(last.lower()+" ")
            g.write(phone+"\n")
            g.write(address.capitalize()+"\n")
# writingboth()
    
```

```python
#//////     program A AND B IST PART
def phoneno():
    with open("phonedir1.txt","r") as f:
        search=input("\nENTER FIRST NAME: ")+" "+input("ENTER LAST NAME: ")
        search=search.Capitalize()
        flag=False
        for x in f:
            y=x.rstrip("\n")
            list1=y.split()
            name1=list1[0]+" "+list1[1]  
            if search == name1:
                print("--------------------------------------------------------")
                print(f"\n^-^ The PHONE NUMBER Of {search} Is:-> {list1[-1]}")
                print("\n--------------------------------------------------------")
                flag=True
                break
        if flag==False:
            print("--------------------------------------------------------")
            print("\n--> NO SUCH PERSON NAME IS PRESENT IN FILE.")
            print("\n--------------------------------------------------------")
        repeat()        
```

```python
def repeat():
    another=input("\nDO YOU WANT TO REPEAT-> YES OR NO: ")
#     another=input()
    if another=="yes" or another=="YES":
        phoneno()
    elif another=="no" or another=="NO":
        print("--------------------------------------------------------")
        print("\n^-^ THANK FOR USING THIS.\n")
        print("--------------------------------------------------------")
    else:
        while True:
            print("\nEnter Correct Command.")
            another=input("YES OR NO")
            if another=="yes" or another=="YES":
                phoneno()
                break
            elif another=="no" or another=="NO":
                print("--------------------------------------------------------")
                print("\n^-^ THANK FOR USING THIS.\n")
                print("--------------------------------------------------------")
                break
            else:
                continue
```
