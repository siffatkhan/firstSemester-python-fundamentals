# 2. ENCRYPTION PROGRAM

```python
#2. ENCRYPTION PROGRAM          

data=input("Enter four digit integer: ")

if len(data)==4:
    index0=int(data[0])
    index0+=7
    index0a=index0%10

    index1=int(data[1])
    index1+=7
    index1a=index1%10

    index2=int(data[2])
    index2+=7
    index2a=index2%10

    index3=int(data[3])
    index3+=7
    index3a=index3%10

    x=index0a                     #HERE INTERCHANGED THE VALUE OF INDEX 0 WITH 2
    index0a=index2a
    index2a=x

    y=index1a                     #HERE INTERCHANGED THE VALUE OF INDEX 1 WITH 3
    index1a=index3a
    index3a=y
    
    print()
    print("|---------------------------------|")
    print("| ENCRYPTED INTEGER IS :",index0a,index1a,index2a,index3a," |")
    print("|---------------------------------|")

else:
    print("\nEnter 4 Valid Integers.")
#   raise ValueError("enter valid iintegers")
```

# DECRYPTION PROGRAM

```python
#2. DECRYPTION PROGRAM

data=input("Enter four digit integer: ")
   

if len(data)==4:
    index0=int(data[0])              # value=(quotient*divisor)+remainder
    index0-=7
    index0=(index0%10)

    index1=int(data[1])           
    index1-=7
    index1=(index1%10)   

    index2=int(data[2])           
    index2-=7
    index2=(index2%10)

    index3=int(data[3])           
    index3-=7
    index3=(index3%10)

    
    x=index0                        #HERE INTERCHANGED THE VALUE OF INDEX 0 WITH 2
    index0=index2
    index2=x
    
    y=index1                        #HERE INTERCHANGED THE VALUE OF INDEX 1 WITH 3
    index1=index3
    index3=y
    
    print()
    print("|---------------------------------|")
    print("| DECRYPTED INTEGER IS :",index0,index1,index2,index3," |")
    print("|---------------------------------|")
elif len(data)!=4:
    print("\nEnter 4 Valid Integers .")
#   raise ValueError("enter valid iintegers")               #WE CAN ALSO RAISE CUSTOM ERROR HERE
```

# 3. WORLD POPULATION GROWTH

```python
#3. World Population Growth

population=float(input("Enter approx current world population: "))
growth_rate=float(input("Enter approx current population growth rate: "))
print()
print("-----------------------------------------------------------------------------------------------")
print("YEAR\t  POPULATION\t\t\t\t  INCREMENT\t\t\t  ")
for i in range(1,75+1):
    if i==1:
        print("-----------------------------------------------------------------------------------------------")
        print(f" {i}\t  {population}\t\t\t\t   {0}\t\t\t\t  ")
    else:
        x=((growth_rate/100)*population)            
        population+=x                               

        print("----------------------------------------------------------------------------------------------")
        print(f" {i}\t  {round(population,2)}\t\t\t\t   {x:.2f}\t\t\t  ")
print("-------------------------------------------------------------------------------------------------")
print()

double_population = population * 2
for z in range(1, 75+1):
    x = population * (growth_rate/100)
    population += x
    if population == double_population:
        print("The year in which the population will become double or get more is: ",z)
    else:
        print("Exact Doubling Of Current Population Is Not Possible ")
        print()
        break
```

# 4.GAME OF CHANCE

```python
import random
balance=int(input("Enter Balance: "))
while True:
    num=random.randint(0,30)
    amount=int(input("\nEnter Betting Amount Between 1-10: "))
    balance-=amount
    if amount <1 or amount>10:
        print("\nBetting should be between 1-10: ")
        break

    guess=int(input("Enter Your Guess From 1-30: "))
    if guess < 0 or guess>30:
        print("\nInvalid guess.")
        break
        
    if guess==num:
        print("========================================================")
        if num==0 or num==1:
            None
        else:
            if num%2==0:
                amount*=2
                print("Congrats Its Even Number, U Got 2X Award.")
                print("--------------------------------------------------------")
                balance+=amount  
        
        if num==0:
            None
        else:
            if num%10==0:
                amount*=3
                print("Congrats Its Multiple Of 10, U Got 3X Award.")
                print("--------------------------------------------------------")
                balance+=amount
    
        if num<5:
            amount*=2
            print("Congrats Its Below 5 Number, U Got 2X Award.")
            print("--------------------------------------------------------")
            balance+=amount
        
        count=2
        prime=True
        if num==0 or num==1:
            None
        else:
            while count<num:
                if num%count==0:
                    prime=False
                    break
            
                count+=1
            if prime==True:
                amount*=5
                balance+=amount
                print("Congrats Its Prime Number,U Get 5X Award.")
                print("--------------------------------------------------------")
                
        print("Total Payout Amount",balance)
        print("========================================================")      
         
    else:
        print("========================================================")
        print("Incorrect Guess")
        print("========================================================")
    print("Total Balance Left: ",balance)
    print("========================================================")
    if balance<=0:
        print("========================================================")
        print("\nSorry you have no money left for betting.")
        print("Add more betting money")
        balance=int(input("Enter Balance: "))
        print("========================================================")
    
    loop=input("\nDou You Want To Play Again.\n")
    if loop =="yes" or loop =="YES":      
        continue
    else:
        break
```
