```python
number=int(input("enter a number: "))    #IRA METHOD
def main():
    z=isprime(number)
    if z:
        print("Prime number")
    else:
        print("composite number")
        
def isprime(n):
    prime=True
    for x in range(2,n):
        if n%x==0:
            prime=False
#             break
        return prime        

main()
```

```python
r1=int(input("enter starting point"))
r2=int(input("enter last point"))

for x in range(r1,r2+1):


    count=2
    prime=True
    while count<x:
        if x%count==0:
            prime=False
            break
        count+=1
    if prime==True:
        print(x,"is a prime number")
        
    else:
        print(x,"is a composite number")
```

```python
r1 = int(input("Enter Starting number: "))
r2 = int(input("Enter 2nd number: "))

print("Prime numbers between", r1, "and", r2)

if r1 < 2:
    r1 = 2  # Ensuring we start from a valid prime candidate

while r1 < r2:
    prime = True
    count = 2
    while count * count <= r1:  # Checking only up to sqrt(r1)
        if r1 % count == 0:
            prime = False
            break
        count += 1

    if prime:  # Print only if prime remains True
        print(r1)

    r1 += 1  # Ensure increment is outside the inner loop
```
